# The special theory of relativity

We are searching for a PGA in which the speed of light must be the same in every inertial reference frame. 
<!-- For simplicity, let us consider only the $(x, t)$ plane. Then an event is given by $$  -->

Let $S$ and $S'$ be two inertial frames, with the origin of $S'$ moving with velocity $v_x$ in the $x$ direction as observed from $S$.



Then relativity demands that for a lightray observed in either $S$ or $S'$, we have 
$$\Delta x = c \Delta t \qquad \Delta x' = c \Delta t' \ ,$$
where $c$ is the speed of light.


Consequentualy, we have that the following quantity is also common to both frames:
$$ (\Delta x)^2 - c^2 (\Delta t)^2 = (\Delta x')^2 - c^2 (\Delta t')^2 = 0$$

# The special theory of relativity

For 3D space we used the algebra $\mathbb{R}_{3,0,1}$. 
So if we want to include *time* in our algebra, we need to add an extra dimension to represent the time dimension, in such a way that
$$ (\Delta x)^2 - c^2 (\Delta t)^2 = (\Delta x')^2 - c^2 (\Delta t')^2 = 0 \ .$$

For the extra basis vector $\mathbf{e}_t$ there are only three options: $\mathbf{e}_t^2 = +1$, $\mathbf{e}_t^2 = 0$, $\mathbf{e}_t^2 = -1$.

Singling out the $x, t$ plane for a moment, this means we have
$$ \ell = \Delta x \mathbf{e}_x + c \Delta t \mathbf{e}_t$$
and thus
$$ 0 = \ell^2 = (\Delta x)^2 \mathbf{e}_x^2 + c^2 (\Delta t)^2 \mathbf{e}_t^2 $$
demands that $\boxed{\mathbf{e}_x^2 = - \mathbf{e}_t^2}$.

# The special theory of relativity

Let's also verify that light rays are invariant under coordinate transformations, 
really ensuring that the second postulate holds.

Up to scaling, lightrays are given by $\ell_{\pm} = (\mathbf{e}_x \pm \mathbf{e}_t)$.
Coordinate transformations in the $(x, t)$-plane are generated by the bivector $\mathbf{e}_{xt}$:
$$ \Lambda = e^{\omega \mathbf{e}_{xt}} = \cosh(\omega) + \sinh(\omega) \mathbf{e}_{xt}$$
It is easy to verify that $\ell_{\pm}$ are eigenvectors of $\Lambda$ and thus satisfy
$$ \Lambda \ell_{\pm} \widetilde{\Lambda} = \pm \ell_\pm $$

# Principle of Covariance

The **principle of covariance** emphasizes that physical laws should be formulated using only quantities that can be unambiguisly correlated by observers in different frames of reference.

Hence, mathematically physical quantities must transform **covariantly** under coordinate transformations.

Let $\circ$ be a binary operation in PGA. We say $\circ$ transforms covariantly if for a transformation $R \in \text{Pin}(p,q,r)$ we have
    $$ R[a \circ b] = R[a] \circ R[b] $$
All of PGA's binary operators $(*, \, \cdot,\, \wedge, \vee, +, -)$ satisfy this property, making PGA a very expressive framework in which to formulate natural laws.

Imagine Alice is standing still, while Bob is on a passing train. As Irresponsible Bob reaches Alice at $t=0$, he fires a gun in the direction of the train.

The train is moving at velocity $v_\text{train}$, while the bullet is traveling with a velocity $v_\text{bullet}$ as seen by Bob. At what velocity does Alice see the bullet moving?

$$ v_{\text{bullet}}' = v_\text{train} + v_\text{bullet} $$

# ~~Addition of Velocity~~ â†’ Multiplication of Rotors!

Let's look at this again from a group-theoretical perspective. Bob is moving with a rotor $R_\text{Bob} = e^{-\tfrac{1}{2} v_\text{Bob} t}$ **as seen by Alice**, while the bullet travels with $R_\text{bullet} = e^{-\tfrac{1}{2} v_\text{bullet}t}$ **as seen by Bob**.

Thus Alice sees the bullet being translated by the motor
$$ R_\text{bullet} = R_\text{bullet} R_\text{Bob} = e^{-\tfrac{1}{2} (v_\text{bullet} + v_\text{Bob})t} $$
So we find the same result: $v_\text{bullet}' = v_\text{bullet} + v_\text{Bob}$.

# Relativistic Addition of Velocity

Bob did very well with his plumbing business, and is now cruising around in his spaceship.
Upon passing Alice with $v = 0.9c$, Bob fires a torpedo with $u' = 0.87c$. 
What is the velocity $u$ of the torpedo as seen by Alice?

Clearly the answer is not $u = v + u' = 1.77c$, as this would violate the postulates of special relativity.
Textbooks will tell you to instead use
$$ u = \frac{v + u'}{1 + v u'} \approx  0.99 c$$

You might be thinking: isn't this a mess? And this time you'd be right!

# ~~Relativistic Addition of Velocity~~ â†’ Multiplication of Rotors!

Boosts are of the form 
$$ \Lambda(\omega) = e^{\omega} = c(\omega) + s(\omega) = c(\omega) (1 + t(\omega)), $$
where $\omega = \vec{\omega} \gamma_0 $ is the rapidity bivector.

If we define the Lorentz factor $\gamma = c(\omega)$ and $v/c = t(\omega)$, then this can also be written as
$$ \Lambda(v) = \gamma (1 + \frac{v}{c}). $$

The rotor approach works in both cases, and is therefore far more intuitive.

# ~~Relativistic Addition of Velocity~~ â†’ Multiplication of Rotors!

For a boosts of the form 
$$ \Lambda(\omega) = e^{\omega} = c(\omega) + s(\omega) = c(\omega) (1 + t(\omega)), $$
where $\omega = \vec{\omega} \gamma_0 $ is the rapidity bivector, we have
$$ \gamma = \langle \Lambda \rangle , \quad \beta = c v = \langle \Lambda \rangle_2 / \langle \Lambda \rangle $$

This also means that in order to compute the Lorentz factor $\gamma$ between two worldlines $\ell_1$ and $\ell_2$ as seen from any frame, is always $\gamma_{12} = \ell_1 \cdot \widetilde{\ell_2}$, while $v_{12} = \ell_1 \times \widetilde{\ell_2}$.
