## An Algebra of Geometry

So far we have only discussed geometry with oriented planes, but done no algebra.
<div class="fragment">
But the rules I have explained are exactly the rules of Geometric Algebra (GA):
a Geometric (Clifford) Algebra $\mathbb{R}_{pqr}$ of $d$ = $p+q+r$ dimensions has $d$ anti-commuting ($\mathbf{e}_i \mathbf{e}_j=-\mathbf{e}_j \mathbf{e}_i$) basis vectors $\mathbf{e}_i$, of which $p$ are positive basis vectors $\mathbf{e}_i^2 = 1$, $q$ are negative basis vectors $\mathbf{e}_i^2 = -1$, and $r$ are null basis vectors $\mathbf{e}_i^2 = 0$.
</div>

<div class="fragment">
To represent Euclidean geometry in $n$ dimension, use $\mathbb{R}_{n,0,1}$:
    <ul>
        <li class="fragment">$n$ basis vector $\mathbf{e}_i$ with $\mathbf{e}_i^2 = 1$ to represent orthogonal (hyper)planes through the origin</li>
        <li class="fragment">$\mathbf{e}_0$ with $\mathbf{e}_0^2 = 0$ to represent the (hyper)plane at infinity.</li>
    </ul>
</div>
<div class="fragment">
Now the equation of a hyperplane in $\mathbb{R}^{n}$, given by $ a_1 x_1 + a_2 x_2 + \ldots + a_n x_n = -b$, is represented as a vector $a$ in the GA $\mathbb{R}_{n,0,1}$:
<div style="display: flex; gap: 1em; align-items: flex-start;">
  <div data-markdown style="flex: 1;">
$$ a = a_1 \mathbf{e}_1 + a_2 \mathbf{e}_2 + \ldots + a_n \mathbf{e}_n + b \mathbf{e}_0 \ .$$
  </div>
</div>
</div>

## The special theory of relativity

When things go very fast, Newtonian mechanics is no longer valid.
The special theory of relativity by Einstein adresses the shortcomings of Newtonian mechanics, by starting from two simple postulates with far reaching consequences:

- **Postulate 1**: (Galilean) Principle of relativity: the laws of physics are the same in all
inertial frames of reference.
- **Postulate 2**: the speed of light in vacuum is the same in all inertial frames of reference.

<div style="text-align: center;">What led  Einstein to introduce this drastic second postulate?</div>
$$ 
\partial F = J \qquad \begin{cases}
\nabla \cdot \mathbf{E} &= \rho \qquad \nabla \times \mathbf E = -\frac{1}{c} \frac{\partial \mathbf B}{\partial t} \\
\nabla \cdot \mathbf B &= 0 \qquad \nabla \times \mathbf B = \frac{1}{c} (\mathbf J + \frac{\partial \mathbf E}{\partial t})
\end{cases}
$$
<div style="text-align: center;">
Note how the speed of light $c = 299 \ 792 \ 458 \ m s^{-1}$ features in Maxwell's equations as a constant.
</div>
<div data-markdown class="fragment r-stack r-frame" style="text-align: center;">
<span class="fragment current-visible">Either Maxwell's equation is not a Law of nature, or the speed of light is the same for every observer.</span>
<span class="fragment current-visible">
The Michelson–Morley experiment (1887) showed for the first time that the speed of light is independent of the velocity of the observer.
</span>
<span class="fragment current-visible">
Ergo: the second postulate cannot be avoided. Consequence: distance and time are relative!
</span>

## The special theory of relativity

<div style="text-align: justify;">
We are searching for a PGA where the speed of light is the same in every inertial reference frame. 
</div>
<div style="text-align: justify;" class="fragment">
Let $S$ and $S'$ be two inertial frames, with the origin of $S'$ moving with velocity $v_x \neq 0$ in the $+x$ direction as observed from $S$.
</div>
<div style="text-align: justify;" class="fragment">
Recall that velocity is defined as $ v_x = \frac{\Delta x}{\Delta t}$.
</div>
<div style="text-align: justify;" class="fragment">
Then the second postulate demands that for a lightray observed in either $S$ or $S'$, we have 
$$\Delta x = c \Delta t \qquad \Delta x' = c \Delta t' \ ,$$
where $c$ is the speed of light.
</div>


Consequentualy, we have that the following quantity is also common to both frames:
$$ (\Delta x)^2 - c^2 (\Delta t)^2 = (\Delta x')^2 - c^2 (\Delta t')^2 = 0$$

## The special theory of relativity

So if we want to include *time* in our algebra, we need to add an extra time dimension such that for light
$$ (\Delta x)^2 - c^2 (\Delta t)^2 = (\Delta x')^2 - c^2 (\Delta t')^2 = 0  \ .$$

For the extra basis vector $\mathbf{e}_t$ there are only three options: $\mathbf{e}_t^2 = +1$, $\mathbf{e}_t^2 = 0$, $\mathbf{e}_t^2 = -1$.

For a lightray in the $(x, t)$ plane, this means we have
$$ \ell = \Delta x \mathbf{e}_x + c \Delta t \mathbf{e}_t$$
and thus
$$ 0 = \ell^2 = (\Delta x)^2 \mathbf{e}_x^2 + c^2 (\Delta t)^2 \mathbf{e}_t^2 $$
demands that $\boxed{\mathbf{e}_x^2 = - \mathbf{e}_t^2}$.

For the remainder of this talk, the Spacetime Algebra Projectivized (STAP) is $\mathbb{R}_{n,1,1}$.
</div>

## The problem with simultaneity

<div data-markdown class="r-stack" style="text-align: center;">
    <span data-markdown class="fragment current-visible" data-fragment-index="0">An observer $S$ at rest only moves in the time direction.</span>
    <span data-markdown class="fragment current-visible" data-fragment-index="1">An observer $S'$ moving with constant velocity $v_x$ traces out a straight line.</span>
    <span data-markdown class="fragment current-visible" data-fragment-index="2">Now imagine that at $t=0$ in the $S$ frame, we have two events that occur <i>simultaneously</i>:<br>
<div style="float:left">💥</div>🫣<div style="float:right">💥</div>
        <br>Do they occur simultaneously as observed from the $S'$ frame?
    </span>
    <span data-markdown class="fragment current-visible" data-fragment-index="3">To answer this question, we have two options:<ol>
  <li>Draw the grid lines of the $S'$ frame as observed from the $S$ frame.</li>
  <li>Boost to the $S'$ frame so we the events from their perspective.</li>
</ol></span>
    <span data-markdown class="fragment current-visible" data-fragment-index="4">1. Draw the grid lines of the $S'$ as observed from the $S$ frame.<br>
        Boost the grid of $S$ to the $S'$ frame using $\Lambda_{1/2} = \sqrt{\ell' / \ell} = \overline{(\ell + \ell')} / \ell$.
    </span>
    <span data-markdown class="fragment current-visible" data-fragment-index="5">We see that for the observer $S'$, event 1 happens well before event 2!</span>
    <span data-markdown class="fragment current-visible" data-fragment-index="6">2. Boost to the $S'$ frame so we the events from their perspective.<br>
        Align the wordline $\ell'$ with $\ell$ by applying $\Lambda_{1/2}^{-1} = \sqrt{\ell' / \ell'} = \overline{(\ell + \ell')} / \ell'$ to the entire diagram.
    </span>
    <span data-markdown class="fragment current-visible" data-fragment-index="7">The $S'$ perspective makes it even clearer that from their perspective, 1 happens before 2.</span>
</div>

## Boosts

How do we compute the boost from one frame to the other?
<span>
We already saw geometrically that the answer is $\Lambda_{1/2} = \sqrt{\ell' / \ell}$.
</span>
<span>
However, typically we are given a velocity, e.g. that a muon travels at a speed $v_x \approx 0.99 c$. In that case, how do we compute $\Lambda$ directly?
</span>
<ol class="fragment">
    <li>Define the dimensionless quantity $\beta_x = v_x / c$</li>
    <li class="fragment">Form $L = 1 - \beta_x \mathbf{e}_{xt}$ just like we would in Euclidean PGA to create a translation.</li>
    <li class="fragment">A bireflection should be normalized, so let's check that: $$L \widetilde{L} = (1 - \beta_x \mathbf{e}_{xt}) (1 + \beta_x \mathbf{e}_{xt})= (1 - \beta_x^2 \mathbf{e}_{xt}^2)$$ and $\mathbf{e}_{xt}^2 = +1$.</li>
    <li class="fragment">So the bireflection we are looking for is $\Lambda = \overline{L} = \overline{1 - \beta_x \mathbf{e}_{xt}}$ where overline means "normalized".</li>
</ol>

<div class="fragment">
Call the normalization factor $\gamma$ such that $\Lambda = \gamma L$. 
<span class="fragment">Then:
$$ 
\begin{aligned}
1 = \Lambda\widetilde{\Lambda} = \gamma^2 L \widetilde{L} &= \gamma^2  (1 - \beta_x \mathbf{e}_{xt}) (1 + \beta_x \mathbf{e}_{xt})= \gamma^2 (1 - \beta_x^2 \mathbf{e}_{xt}^2)
\end{aligned}
$$
</span>
<span class="fragment">
Since $\mathbf{e}_{xt}^2 = +1$ in STA, we obtain $\gamma = \sqrt{\frac{1}{1 - \beta^2}}$. This quantity is known as the <i>Lorentz factor</i>.
</span>
</div>

<div class="fragment"  style="text-align: center;">
    $$ \boxed{\Lambda_{1/2} = \sqrt{\Lambda} = \overline{1 + \Lambda}} $$
</div>
<span class="fragment">
A large part of a young relativists live is spend computing Lorentz factors. 
</span>

Consider two clock ticks that occur an amount $T$ apart in the $S'$ frame. What is the time $T'$ between them in the $S$ frame?

<div data-markdown class="r-stack" style="text-align: center;">
<div class="fragment current-visible">Boost to the $S$ frame with $\Lambda_{1/2}$.</div>
<div class="fragment current-visible">Project the events $E_i'$ onto $\ell$ with $(E_i' \cdot \ell)/\ell$.</div>
<div class="fragment current-visible">We see that $T = \gamma T'$, a phenomenon known as <i>time dilation</i>.</div>
<div class="fragment current-visible"><b>Consequence</b>: because the gaps between ticks on the moving clocks take longer, <br>moving clocks appear to run slower!</div>

<div class="fragment current-visible"><i>Answer:</i> magnitude of the line $[(E_1' \cdot \ell)/\ell] \vee [(E_2' \cdot \ell)/\ell] = ((E_1' \vee E_2') \cdot \ell)/\ell$</div>
</div>

### Example: muon decay

Muons are the heavier and unstable version of the electron, with a half-life of $\tau \approx 2 \times 10^{-6}$. These particles are created when cosmic rays hit the upper atmosphere, and it takes them $t \approx 6 \times 10^{-6}$ to get to sea level. Despite this being more than three times their half-life, large quantities of them make it down to earth. But because they travel at $v \approx 0.99 c$, corresponding to $\gamma \approx 10$, their half-life is $\approx 2 \times 10^{-5}$ as viewed from the earth. Plenty of time for them to make it!

# Natural Laws

We have been sidetracked by thinking about things which are different for different observers, but our original goal was to formulate the **laws of nature** in a way that looks the same for all observers.
The **principle of covariance** emphasizes that physical laws should be formulated using only quantities that can be unambiguisly correlated by observers in different frames of reference.

Under $R \in \text{Pin}(p,q,r)$ the binary operators $\circ \in (\;\;, \, \cdot,\, \wedge, \vee, +, -)$ of PGA transform as:
    $$ R[a \circ b] = R[a] \circ R[b] $$
This makes PGA a very expressive framework in which to formulate natural laws.

## Relativistic Kinematics with Geometric Operators

A point particle can be modeled by an event that evolves under a motor $M(\tau) \in \text{Pin}(n, 1, 1)$:
$$ X(\tau) = M X_0 \widetilde{M} \ , $$
where $X_0$ is the initial position at $\tau=0$.
<span>
The derivative of $X$ is given by 
$$ \dot X(\tau) = X(\tau) \times B(\tau) \quad \text{where} \quad B = - 2 \dot M \widetilde{M} \ .$$
</span>

# Final Thoughts

- We formulated an Algebra for Minkowski space, a.k.a. Spacetime Algebra (STA), whose isometry group is the (double cover) of the Lorentz group $\text{Pin}(3,1)$.
- By including the hyperplane at infinity this was extended to the Spacetime Algebra Projectivized (STAP), whose isometry group is the (double cover) of the Poincaré group $\text{Pin}(3,1,1)$.
- All the binary operators of STAP transform covariantly with Poincaré transformatons and thus allow us to write down the laws of nature in a way that complies with the principle of relativity.
